```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Math, display
```

```python
X = np.arange(1, 11)
Y = abs(np.array([-2.9443,
-5.965869,
-8.988,
-12.009,
-15.031,
-18.054,
-20.999,
-23.944,
-27.045,
-30.035])) * 1e-3
L = 1
b = 0.15 * 1e-3
```

```python
X_square = X*X
Y_square = Y*Y
XY = X*Y
```

```python
tg_phi, A = np.polyfit(X, Y, deg=1)
f = np.poly1d([tg_phi, A])
display(Math(f'$\\tg{{\\phi}} = {tg_phi}$'))

plt.plot(X, Y, 'x-', label="Experemental")
plt.plot(X, f(X), label="Trend")
plt.text(1, 0.03, f'$y = {tg_phi:.2e}x + {A:.2e}$')
plt.xlabel('$m$')
plt.ylabel('$x_m [m]$')
plt.legend()
plt.show()
```

```python
mean_lambda_graph = b*tg_phi/L
display(Math(f'$\\langle \\lambda_{{graph}} \\rangle = {round(mean_lambda_graph * 1e9)} \\cdot 10^{{-9}}[m]$'))
```

```python
B = (XY.mean() - X.mean() * Y.mean()) / (X_square.mean() - X.mean() ** 2)
S_b = 1 / np.sqrt(X.size - 2) * np.sqrt((Y_square.mean() - Y.mean() ** 2) / (X_square.mean() - X.mean() ** 2) - B*B)
display(Math(f'$B = {B:.2e}$'))
display(Math(f'$S_B = {S_b:.2e}$'))
```

```python
mean_lambda_analyt = b*B/L
display(Math(f'$\\langle \\lambda_{{analyt}} \\rangle = ({round(mean_lambda_analyt * 1e9, 1)} \\pm {round(mean_lambda_analyt * S_b / B * 1e9, 1)}) \\cdot 10^{{-9}} [m]$'))
```
