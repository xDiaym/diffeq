```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 1e-2
x0, xn = 0, 1
```

```python
def f(x: float, iters: int = 100) -> float:
    y = np.zeros_like(x)
    for n in range(1, iters):
        y += 2 * (-1) ** (n + 1) / (np.pi * n) * np.sin(np.pi * n * x)
    return y
```

```python
X = np.arange(x0, xn, dx)
Y = f(X)
```

```python
plt.plot(X, Y, label='$S_n(x)$')
plt.plot(X, X, label='$x$')
plt.legend()
plt.show()
```
