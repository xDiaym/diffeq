```python
import numpy as np
import matplotlib.pyplot as plt

C = np.array([-40, -10, 0, 8, 15, 22, 38])
F = np.array([-40, 14, 32, 46, 59, 72, 100])
B = np.ones(F.shape)
lmb = 0.001
def f(x):
    return x

w = np.array([1, 1])
MSE = []
count = []

for i in range(10000):
    k = np.random.randint(0, len(C))
    X = np.array([C[k], B[k]])
    y = np.dot(X, w)
    out = f(y)

    e = out - F[k]
    MSE.append(0.5 * e**2)
    count.append(i)
    w = w - lmb*1*e*X

print(w)
plt.plot(count, MSE)
plt.show()
```
