# Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x <= 0).astype(np.float64)

W1 = np.array([[-4, 4, -1], [4, -4, -1], [-2, -2, 1], [2/3, 2/3, -1]])
W2 = np.array([-1/4, -1/4, -1/4, -1/4, 1])

def perceptron(x):
    z1 = f(W1 @ np.concatenate([x, np.ones(1)]).T)
    z2 = f(W2 @ np.concatenate((z1, np.ones(1))).T)
    return z2
```

```python
np.random.seed(42)
X = np.random.random((2500, 2))
Y = np.array([perceptron(x) for x in X])
red = X[Y == 0]
green = X[Y == 1]

plt.scatter(red[:, 0], red[:, 1], c='red', s=15)
plt.scatter(green[:, 0], green[:, 1], c='green', s=15)
plt.plot([0.125, 0.625, 0.875, 0.375, 0.125], [0.375, 0.875, 0.625, 0.125, 0.375])
plt.show()
```
