```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [lambda x: 1, lambda x: -2, lambda x: 1]
dx = 1e-2
x_0, y_0 = 0, 1
x_n, y_n = 3, 0.5*np.exp(3)
n = int((x_n - x_0) / dx)
```

```python
x, alpha, beta = x_0, [0], [y_0]
A = lambda x: a[0](x)/dx**2 + a[1](x)/dx
B = lambda x: -2*a[0](x)/dx**2 -a[1](x)/dx + a[2](x)
C = lambda x: a[0](x)/dx**2
for i in range(1, n):
    d = B(x+dx) + C(x+dx)*alpha[-1]
    alpha.append(-A(x+dx) / d)
    beta.append(-C(x+dx) * beta[-1] / d)
    x += dx
```

```python
plt.plot(np.arange(x_0, x_n, dx), alpha, label=r'$\alpha$')
plt.plot(np.arange(x_0, x_n, dx), beta, label=r'$\beta$')
plt.legend()
plt.show()
```

```python
y = [0] * (n-1) + [y_n]
for i in range(n-1, 0, -1):
    y[i-1] = (alpha[i]*y[i]+beta[i])
```

```python
space = np.arange(x_0, x_n, dx)
plt.plot(space, y, label='approx')
plt.plot(space, (1-1/6*space)*np.exp(space), label='exact')
plt.legend()
plt.show()
```
